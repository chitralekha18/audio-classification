```python
import json
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
```

```python
# load the json file
f = open('nsjson_softlabels_mosquitobrass.json')
data = json.load(f)
```

```python
data
```

```python
# create the list for visualization
# softlabel_3_2_(0-2) --> the 3 penultimate layer node values
# softlabel_3_2_(3-4) --> the 2 final layer node values
xaxis = []
yaxis = []
zaxis = []
node0 = []
node1 = []
label = [] # 0 for water, 1 for wind


for i in data:
    xaxis.append(data[i]['softlabel_3_2_0'])
    yaxis.append(data[i]['softlabel_3_2_1'])
    zaxis.append(data[i]['softlabel_3_2_2'])
    node0.append(data[i]['softlabel_3_2_3'])
    node1.append(data[i]['softlabel_3_2_4'])
```

```python
# identify the true class of the sound file
for j in data:
    if 'Brass' not in data[j]['sound_name']:
        label.append(0)
    else:
        label.append(1)       
```

```python
label
```

```python
# plot the soft labels for each sound file
plt.figure()
ax = plt.axes(projection='3d')
scatter = ax.scatter3D(xaxis, yaxis, zaxis,c=label, alpha=0.5, s=20)
#ax.set_xlim3d(0, 0.004)
# ax.set_ylim3d(0.8, 1)
# ax.set_zlim3d(0, 0.05)
plt.xlabel('dim 0')
plt.ylabel('dim 1')
ax.set_zlabel('dim 2')
plt.title('Penultimate layer outputs of (3_2)')
```
